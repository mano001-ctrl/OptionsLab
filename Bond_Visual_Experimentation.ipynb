```python
import pandas as pd

# Sample data structure
data = {
    'Issuer': ['Federal Government of Nigeria', 'Afreximbank', 'Absa Group Limited', 'Republic of South Africa'],
    'Coupon Rate (%)': [14.8, 2.634, 6.375, 'Varies'],
    'Maturity Date': ['2049-04-26', '2026-05-17', 'Perpetual', 'Varies'],
    'Currency': ['NGN', 'USD', 'USD', 'ZAR'],
    'Credit Rating': ['B-', 'A-', 'BB+', 'BB-'],
    'Yield to Maturity (%)': [None, None, None, None],  # Replace None with actual data
    'Liquidity': ['Moderate', 'High', 'High', 'High']
}

df = pd.DataFrame(data)
```

```python
import matplotlib.pyplot as plt

# Filter out bonds with varying coupon rates
filtered_df = df[df['Coupon Rate (%)'] != 'Varies']

plt.figure(figsize=(10, 6))
plt.bar(filtered_df['Issuer'], filtered_df['Coupon Rate (%)'], color='skyblue')
plt.xlabel('Issuer')
plt.ylabel('Coupon Rate (%)')
plt.title('Comparison of Coupon Rates Among Bonds')
plt.xticks(rotation=45)
plt.show()
```

```python
# Mapping credit ratings to numerical values
rating_map = {'AAA': 1, 'AA+': 2, 'AA': 3, 'AA-': 4, 'A+': 5, 'A': 6, 'A-': 7,
              'BBB+': 8, 'BBB': 9, 'BBB-': 10, 'BB+': 11, 'BB': 12, 'BB-': 13,
              'B+': 14, 'B': 15, 'B-': 16}

df['Credit Rating Numeric'] = df['Credit Rating'].map(rating_map)

plt.figure(figsize=(10, 6))
plt.scatter(df['Credit Rating Numeric'], df['Yield to Maturity (%)'], color='green')
plt.xlabel('Credit Rating (Numeric)')
plt.ylabel('Yield to Maturity (%)')
plt.title('Credit Rating vs. Yield to Maturity')
plt.xticks(ticks=list(rating_map.values()), labels=list(rating_map.keys()), rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the radar chart
categories = ['Coupon Rate', 'Liquidity', 'Risk']

# Add values for all bonds mentioned in the conversation
bond_data = {
    'FGN 14.8% Apr 2049': [14.8, 6, 8],
    'R186': [10.5, 9, 6],
    'Afreximbank 2.634% May 2026': [2.634, 7, 5],
    'Absa Group 6.375% Perpetual': [6.375, 8, 6],
    'South African Government Bond (Sample)': [9.0, 8, 5]
}

# Prepare the radar chart data
angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
angles += angles[:1]

# Create the radar chart
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'polar': True})

for bond, values in bond_data.items():
    values += values[:1]  # Close the radar chart
    ax.plot(angles, values, label=bond)
    ax.fill(angles, values, alpha=0.25)

# Configure chart
ax.set_yticks([5, 10, 15])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)
ax.set_title('Comparison of Bonds Across Key Metrics', va='bottom')
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
plt.show()
```
