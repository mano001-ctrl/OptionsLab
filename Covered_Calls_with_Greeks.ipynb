```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the covered call
stock_price = np.linspace(0, 200, 100)  # Stock price range
strike_price = 100  # Strike price of the call option
premium = 10  # Premium received for the call option
initial_stock_price = 100  # Initial stock price

# Payoff of the stock
stock_payoff = stock_price - initial_stock_price

# Payoff of the short call option
call_payoff = np.where(stock_price > strike_price, - (stock_price - strike_price) + premium, premium)

# Total payoff of the covered call
covered_call_payoff = stock_payoff + call_payoff

# Enhanced Plot for Covered Call Payoff Diagram
plt.figure(figsize=(12, 8))

# Plot each component with improved visuals
plt.plot(stock_price, stock_payoff, label="Stock Payoff", linestyle="--", linewidth=1.5, color="blue")
plt.plot(stock_price, call_payoff, label="Short Call Payoff", linestyle="--", linewidth=1.5, color="orange")
plt.plot(stock_price, covered_call_payoff, label="Covered Call Payoff", linewidth=2.5, color="green")

# Add horizontal and vertical reference lines
plt.axhline(0, color="black", linewidth=1.0, linestyle="--")
plt.axvline(strike_price, color="red", linewidth=1.0, linestyle="--", label="Strike Price")

# Annotate key points
plt.annotate(f"Premium = {premium}", xy=(strike_price, premium), xytext=(strike_price + 10, premium + 10),
             arrowprops=dict(arrowstyle="->", color="black"), fontsize=10)

plt.annotate("Breakeven Point", xy=(strike_price - premium, 0),
             xytext=(strike_price - premium - 20, -20),
             arrowprops=dict(arrowstyle="->", color="black"), fontsize=10)

# Formatting the plot with title and labels
plt.title("Enhanced Payoff Diagram of a Covered Call", fontsize=16)
plt.xlabel("Stock Price at Expiration", fontsize=14)
plt.ylabel("Payoff", fontsize=14)
plt.legend(fontsize=12)
plt.grid(alpha=0.5)

# Show the enhanced plot
plt.show()

# Define the Greeks for a covered call
delta = np.where(stock_price < strike_price, 1, 0)  # Delta of stock minus delta of the short call
gamma = np.zeros_like(stock_price)  # Gamma is zero for a covered call (no convexity)
theta = np.where(stock_price > strike_price, -0.1, -0.05)  # Approximation of theta decay
vega = np.zeros_like(stock_price)  # Vega is zero since the position is hedged

# Enhanced Plot for Greeks of a Covered Call
fig, axs = plt.subplots(4, 1, figsize=(14, 24), sharex=True)

# Delta
axs[0].plot(stock_price, delta, label="Delta", color="purple", linewidth=2.5)
axs[0].fill_between(stock_price, delta, 0, color="purple", alpha=0.3)
axs[0].set_title("Delta of a Covered Call", fontsize=16)
axs[0].set_ylabel("Delta", fontsize=14)
axs[0].axhline(0, color="black", linestyle="--", linewidth=1)
axs[0].grid(alpha=0.6)
axs[0].legend(fontsize=12)

# Gamma
axs[1].plot(stock_price, gamma, label="Gamma", color="blue", linewidth=2.5)
axs[1].fill_between(stock_price, gamma, 0, color="blue", alpha=0.3)
axs[1].set_title("Gamma of a Covered Call", fontsize=16)
axs[1].set_ylabel("Gamma", fontsize=14)
axs[1].axhline(0, color="black", linestyle="--", linewidth=1)
axs[1].grid(alpha=0.6)
axs[1].legend(fontsize=12)

# Theta
axs[2].plot(stock_price, theta, label="Theta", color="orange", linewidth=2.5)
axs[2].fill_between(stock_price, theta, 0, color="orange", alpha=0.3)
axs[2].set_title("Theta of a Covered Call", fontsize=16)
axs[2].set_ylabel("Theta", fontsize=14)
axs[2].axhline(0, color="black", linestyle="--", linewidth=1)
axs[2].grid(alpha=0.6)
axs[2].legend(fontsize=12)

# Vega
axs[3].plot(stock_price, vega, label="Vega", color="green", linewidth=2.5)
axs[3].fill_between(stock_price, vega, 0, color="green", alpha=0.3)
axs[3].set_title("Vega of a Covered Call", fontsize=16)
axs[3].set_xlabel("Stock Price at Expiration", fontsize=14)
axs[3].set_ylabel("Vega", fontsize=14)
axs[3].axhline(0, color="black", linestyle="--", linewidth=1)
axs[3].grid(alpha=0.6)
axs[3].legend(fontsize=12)

# Adjust spacing for better visual appeal
plt.tight_layout()
plt.show()

"""
Explanations:
1. Enhanced Payoff Diagram:
   - The covered call combines a stock position with a short call option, resulting in a capped upside and limited downside.
   - The chart shows individual payoffs (stock, short call) and the combined payoff.

2. Greeks of a Covered Call:
   - Delta: Shows sensitivity to changes in the stock price, transitioning from 1 to 0 around the strike price.
   - Gamma: Zero throughout, indicating no curvature.
   - Theta: Reflects time decay, more pronounced above the strike price.
   - Vega: Zero because the position is delta-hedged and insensitive to volatility.
"""
```
