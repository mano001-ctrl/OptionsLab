1. Introduction to Gamma Scalping
Gamma scalping is an advanced options trading strategy used by market makers, proprietary traders, and institutional investors to profit from price movements while maintaining a delta-neutral position. It involves dynamically hedging the delta exposure of an options position to capitalize on volatility.

2. Key Concepts in Gamma Scalping
Before diving into the mechanics, let's understand the key concepts:

Delta (Δ): Measures how much an option's price moves in relation to the underlying asset. A delta of 0.50 means the option will move $0.50 for every $1 move in the underlying.
Gamma (Γ): Measures the rate of change of delta. A high gamma means delta changes rapidly, requiring frequent hedging.
Vega (ν): Measures sensitivity to implied volatility. While not directly part of gamma scalping, changes in volatility can impact profits.
Theta (Θ): Measures time decay. Since options lose value as they approach expiration, gamma scalpers must be aware of the impact of theta decay.
3. How Gamma Scalping Works
Gamma scalping involves the following steps:

Establish a Delta-Neutral Position

A trader typically buys or sells options to construct a portfolio where the net delta is close to zero.
This means small price changes in the underlying asset won’t immediately impact the position's value.
Monitor Price Movements and Adjust Hedging

As the underlying asset moves, delta changes due to gamma.
To remain delta-neutral, the trader must adjust the hedge (typically by buying or selling the underlying asset).
Profiting from Volatility

Gamma scalpers make money when price movements are frequent and large.
If the asset moves up, the trader sells shares to reduce delta; if the asset moves down, they buy shares to increase delta.
The trader effectively buys low and sells high in response to price swings.
Managing Costs and Risks

Frequent trading incurs transaction costs, so gamma scalping works best in liquid markets.
Theta decay can erode profits, requiring traders to balance gamma exposure against time decay.
High implied volatility can make long gamma positions expensive.
4. Example of Gamma Scalping

Scenario Setup:
A trader buys an at-the-money (ATM) call option on a stock trading at $100.

The option has a delta of 0.50 and gamma of 0.05.
The trader holds 10 contracts (equivalent to 1,000 shares).


Step 1: Initial Delta-Neutral Position

Delta = 0.50 × 1,000 = 500 → The trader is exposed to 500 long deltas.

To hedge, the trader sells 500 shares to become delta-neutral.

Step 2: Stock Moves Up to $102

Gamma = 0.05 → Delta increases by 0.05 × $2 = 0.10.
New Delta = 0.60 × 1,000 = 600 (since the call gains delta as the stock rises).

The trader is now 100 shares short from being delta-neutral (since they initially hedged at 500).
To rebalance, they sell 100 more shares, bringing the total short position to 600 shares.

Step 3:
Stock Drops to $98

Gamma = 0.05 → Delta decreases by 0.10 (since delta drops as the stock falls).

New Delta = 0.50 × 1,000 = 500.

The trader is now over-hedged by 100 shares (since they had shorted 600 shares).
They buy 100 shares to restore delta-neutrality.

Step 4: Profit from the Scalping Process

When the stock rose, the trader sold 100 shares at $102.

When the stock fell, they bought 100 shares back at $98.

Profit: (102 - 98) × 100 = $400 (ignoring transaction costs).

5. Considerations and Limitations

Best Environment: Works well in high-volatility markets where frequent price swings allow scalping opportunities.

Time Decay (Theta Risk): If the stock does not move much, the trader may lose money due to option decay.
Transaction Costs: Frequent rebalancing requires low-cost execution to remain
profitable.

Liquidity Concerns: Works best in highly liquid assets to minimize slippage.

6. Conclusion
Gamma scalping is a powerful strategy that benefits from price fluctuations while staying delta-neutral. However, it requires careful risk management, cost control, and an understanding of volatility. It is commonly used by market makers and institutional traders to manage options books efficiently.
