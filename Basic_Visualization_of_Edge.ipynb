```python
import matplotlib.pyplot as plt

def plot_straddle_info():
    # Data from the screenshot
    market_price = 11.55
    theoretical_value = 17.89
    edge = 54.9       # in percent
    hist_win_rate = 65  # in percent

    # Prepare data for plotting
    labels = ["Market Price [Debit]", "Theoretical Value"]
    values = [market_price, theoretical_value]

    # Create the bar chart
    fig, ax = plt.subplots(figsize=(6, 4))
    bars = ax.bar(labels, values, color=["#2E86C1", "#28B463"], width=0.5)

    # Add exact value labels on top of each bar
    for bar in bars:
        height = bar.get_height()
        ax.annotate(
            f"{height:.2f}",
            xy=(bar.get_x() + bar.get_width() / 2, height),
            xytext=(0, 3),  # offset in points above the bar
            textcoords="offset points",
            ha="center",
            va="bottom",
            fontsize=10,
            color="black",
        )

    # Title and axis labels
    ax.set_title("Long Straddle - Undervalued Straddle", fontsize=14, fontweight="bold")
    ax.set_ylabel("Price (USD)", fontsize=12)

    # Turn off top and right spines for a cleaner look
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

    # Annotate Edge and Historical Win Rate on the chart
    annotation_text = (
        f"Edge: {edge:.1f}%\n"
        f"Historical Win Rate: {hist_win_rate}%"
    )
    # Positioning the annotation near the top-right
    ax.text(
        1.05, 0.7, annotation_text,
        transform=ax.transAxes,
        fontsize=11,
        verticalalignment="top",
        bbox=dict(boxstyle="round,pad=0.5", fc="#F0F0F0", ec="gray", alpha=0.8)
    )

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    plot_straddle_info()
```
