```python
import yfinance as yf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define the ticker symbol for which you want to fetch options data
ticker_symbol = "QQQ"  # Example: Apple Inc.

# Fetch options data
ticker = yf.Ticker(ticker_symbol)
expirations = ticker.options  # List of expiration dates

# Initialize an empty DataFrame to hold all options data
options_data = []

# Loop through expiration dates to fetch options data
for exp in expirations:
    # Fetch the options chain for the given expiration date
    opt_chain = ticker.option_chain(exp)

    # Combine calls and puts into a single DataFrame with a column for type
    calls = opt_chain.calls.assign(Type='Call', Expiration=exp)
    puts = opt_chain.puts.assign(Type='Put', Expiration=exp)
    options_data.append(pd.concat([calls, puts]))

# Combine all options data into a single DataFrame
df = pd.concat(options_data, ignore_index=True)

# Pivot the data to create a heatmap structure (e.g., Strike Price vs Expiration Date for Open Interest)
heatmap_data = df.pivot_table(
    index='strike',  # Rows = Strike Prices
    columns='Expiration',  # Columns = Expiration Dates
    values='openInterest',  # Values = Open Interest
    aggfunc='sum'  # Sum up open interest across all options of the same strike price and expiration
)

# Plot the heatmap using the "coolwarm" color map
plt.figure(figsize=(14, 8))
sns.heatmap(
    heatmap_data,
    cmap="coolwarm",  # Alternative colormap for intuitive gradients
    linewidths=0.5,
    annot=False,
    cbar_kws={'label': 'Open Interest'}
)

# Customize the plot
plt.title(f'{ticker_symbol} Options Activity Heatmap (Open Interest)')
plt.xlabel('Expiration Date')
plt.ylabel('Strike Price')
plt.tight_layout()

# Show the heatmap
plt.show()
```
