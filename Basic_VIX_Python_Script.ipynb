```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd

# Fetch VIX historical data
vix = yf.download("^VIX", start="2000-01-01", end="2025-12-31")

# Calculate Moving Averages (for trend analysis)
vix["50-day MA"] = vix["Close"].rolling(window=50).mean()
vix["200-day MA"] = vix["Close"].rolling(window=200).mean()

# Plot VIX with Moving Averages
plt.figure(figsize=(12, 6))
plt.plot(vix.index, vix["Close"], label="VIX Close Price", color="blue", alpha=0.7)
plt.plot(vix.index, vix["50-day MA"], label="50-day MA", color="orange", linestyle="dashed")
plt.plot(vix.index, vix["200-day MA"], label="200-day MA", color="red", linestyle="dashed")

# Highlight Major VIX Spikes
spike_dates = ["2008-10-24", "2020-03-16"]
for date in spike_dates:
    plt.axvline(pd.to_datetime(date), color='gray', linestyle="dotted", alpha=0.7)

plt.title("VIX Index with 50-day & 200-day Moving Averages")
plt.xlabel("Year")
plt.ylabel("VIX Level")
plt.legend()
plt.grid(True)
plt.show()
```
