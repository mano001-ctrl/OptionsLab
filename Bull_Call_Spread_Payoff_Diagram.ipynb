```python
import matplotlib.pyplot as plt
import numpy as np

# Parameters for the Bull Call Spread
stock_prices = np.linspace(40, 60, 100)  # Range of stock prices
strike_price_long = 50  # Strike price of the long call
strike_price_short = 55  # Strike price of the short call
premium_long = 3  # Premium paid for the long call
premium_short = 1  # Premium received for the short call

# Payoff for the long call
long_call_payoff = np.maximum(stock_prices - strike_price_long, 0) - premium_long

# Payoff for the short call
short_call_payoff = premium_short - np.maximum(stock_prices - strike_price_short, 0)

# Total payoff for the Bull Call Spread
total_payoff = long_call_payoff + short_call_payoff

# Plotting the payoffs
plt.figure(figsize=(10, 6))
plt.plot(stock_prices, total_payoff, label="Bull Call Spread", color="blue")
plt.axhline(0, color="black", linestyle="--", linewidth=0.7)
plt.axvline(strike_price_long, color="green", linestyle="--", label=f"Long Strike: {strike_price_long}")
plt.axvline(strike_price_short, color="red", linestyle="--", label=f"Short Strike: {strike_price_short}")
plt.title("Bull Call Spread Payoff Diagram")
plt.xlabel("Stock Price at Expiration")
plt.ylabel("Profit/Loss")
plt.legend()
plt.grid()
plt.show()
```
