```python
# Re-importing necessary libraries after environment reset
import matplotlib.pyplot as plt
import numpy as np

# Parameters for the Butterfly Spread
stock_prices = np.linspace(40, 60, 100)  # Range of stock prices
strike_price_low = 45  # Strike price of the lower long call
strike_price_middle = 50  # Strike price of the short calls
strike_price_high = 55  # Strike price of the higher long call
premium_low = 2  # Premium paid for the lower long call
premium_middle = 1.5  # Premium received for each short call
premium_high = 1  # Premium paid for the higher long call

# Payoff for each option
long_call_low = np.maximum(stock_prices - strike_price_low, 0) - premium_low
short_calls_middle = 2 * (premium_middle - np.maximum(stock_prices - strike_price_middle, 0))
long_call_high = np.maximum(stock_prices - strike_price_high, 0) - premium_high

# Total payoff for the Butterfly Spread
total_payoff = long_call_low + short_calls_middle + long_call_high

# Plotting the payoffs
plt.figure(figsize=(10, 6))
plt.plot(stock_prices, total_payoff, label="Butterfly Spread", color="blue")
plt.axhline(0, color="black", linestyle="--", linewidth=0.7)
plt.axvline(strike_price_low, color="green", linestyle="--", label=f"Low Strike: {strike_price_low}")
plt.axvline(strike_price_middle, color="orange", linestyle="--", label=f"Middle Strike: {strike_price_middle}")
plt.axvline(strike_price_high, color="red", linestyle="--", label=f"High Strike: {strike_price_high}")
plt.title("Butterfly Spread Payoff Diagram")
plt.xlabel("Stock Price at Expiration")
plt.ylabel("Profit/Loss")
plt.legend()
plt.grid()
plt.show()
```
